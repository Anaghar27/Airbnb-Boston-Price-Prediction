### Import Statements

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
```

### Load the dataset

```python
calendar = pd.read_csv("../data/calendar.csv")
```

### Preprocessing of data

```python
calendar = calendar.dropna(subset=["price", "date", "listing_id"])
calendar["price"] = calendar["price"].replace(r"[\$,]", "", regex=True).astype(float)
calendar = calendar[calendar["price"] > 0]
calendar["date"] = pd.to_datetime(calendar["date"])
```

```python
top_listings = calendar["listing_id"].value_counts().head(5).index.tolist()
```

```python
os.makedirs("../outputs", exist_ok=True)
forecast_summaries = []
```

### Forecasting

```python
for listing_id in top_listings:
    df_listing = calendar[calendar["listing_id"] == listing_id][["date", "price"]]
    df_prophet = df_listing.rename(columns={"date": "ds", "price": "y"})

    model = Prophet(daily_seasonality=True)
    model.fit(df_prophet)

    future = model.make_future_dataframe(periods=30)
    forecast = model.predict(future)
    forecast["listing_id"] = listing_id

    forecast.to_csv(f"../outputs//forecast_{listing_id}.csv", index=False)

    fig = model.plot(forecast)
    plt.title(f"Price Forecast - Listing ID {listing_id}")
    fig.savefig(f"../outputs/forecast_plot_{listing_id}.png")

    forecast_summaries.append(forecast[["listing_id", "ds", "yhat", "yhat_lower", "yhat_upper"]])
```

```python
combined_forecast = pd.concat(forecast_summaries, ignore_index=True)
combined_forecast.to_csv("../outputs/forecasted_prices.csv", index=False)
```
